```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Generate synthetic data
np.random.seed(42)
num_samples = 100
house_size = np.random.randint(800, 4000, num_samples)
bedrooms = np.random.randint(1, 6, num_samples)
price = 200000 + 150 * house_size + 50000 * bedrooms + np.random.normal(0, 100000, num_samples)

data = pd.DataFrame({'HouseSizeSqft': house_size, 'Bedrooms': bedrooms, 'Price': price})
data.to_csv('E:\Work\Data analysis\Data analysis projects\Linear regression\housing_data.csv')
```

```python
# Check the first few rows of data
print(data.head())

# Plot the data
plt.scatter(data['HouseSizeSqft'], data['Price'])
plt.xlabel('House Size (sqft)')
plt.ylabel('Price')
plt.title('House Prices vs. Size')
plt.show()
```

```python
# Prepare the data
#X = data[['HouseSizeSqft', 'Bedrooms']]
X = data[['HouseSizeSqft']]
Y = data['Price']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state= 42)
```

```python
# Create a linear regression object
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, Y_train)
```

```python
# Make predictions on the test set
Y_pred = model.predict(X_test)

# Calculate mean squared error and R-squared
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test,Y_pred)

print('mse = ', mse)
print('r2 = ', r2)
```
